```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
dt = pd.read_excel('/BIDataset.xlsx')

# Extract columns
sales_list = dt['Sales'].tolist()
Order_ID = dt['Order ID'].tolist()
catagory=dt['Category'].tolist()
print(sales_list)
print(Order_ID)
print(catagory)
# Sort both lists together (if needed) - optional
# Combine and sort by sales
combined = list(zip(Order_ID, sales_list))
combined.sort(key=lambda x: x[1], reverse=True)

# Unzip back
Order_ID_sorted, sales_list_sorted = zip(*combined)

# Plot
plt.figure(figsize=(12, 6))
plt.bar(Order_ID_sorted, sales_list_sorted)
plt.xlabel("Order ID")
plt.ylabel("Sales")
plt.title("Sales by Order ID")
plt.xticks(rotation=90)  # Rotate x-labels if too crowded
plt.tight_layout()
plt.show()
```
